```python
import numpy as np 
import scipy as sp 
import sympy
from matplotlib import pyplot as plt
```

```python
t,e,d = sympy.symbols('t,\epsilon,\delta')
H= sympy.Matrix([[e+d,t],[sympy.conjugate(t),e-d]])
H
```

```python
evals = list(H.eigenvals().keys())
```

```python
eps = 0
t_val = 1
delta_range = np.linspace(-5,5,64)
fig,axs = plt.subplots(2)
sum=np.zeros_like(delta_range)
for for i,eval in enumerate(evals):
    ev = eval.subs({t:t_val,e:eps})
    ax[0].plot(delta_range,sympy.lambdify(d,ev)(delta_range))

fig.show()
```
